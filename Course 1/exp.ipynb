```python
import pandas as pd
import numpy as np
```

```python
grades = pd.Series([90,70,60])
```

```python
%%timeit -n 100

total= 0

for el in grades:
    total += el
    
total/len(grades)


```

```python
%%timeit -n 100

np.mean(grades)
```

```python
%%timeit -n 10

for label, value in grades.items():
    
    grades.at[label] = value+2
```

```python
%%timeit -n 10

grades =+ 2
```

```python
df = pd.read_csv("Course 1/datasets/census.csv")
df.head()
```

```python
%%timeit -n 3

for state in df['STNAME'].unique():
    avg = np.average(df.where(df['STNAME'] == state).dropna()['CENSUS2010POP'])
    
    print('Counties in state ' + state + 
          ' have an average population of ' + str(avg))
```

```python
%%timeit -n 3

for group, frame in df.groupby('STNAME'):
    avg = np.average(frame['CENSUS2010POP'])
    print('Counties in state ' + group + 
          ' have an average population of ' + str(avg))
```
